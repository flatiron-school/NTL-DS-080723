### Task 2: Write a query to return the last names, email addresses and *billing* addresses of the customers who live in Canada.

```python
# call_on_students(1)
pd.read_sql(
    """
    SELECT c.LastName, c.Email, i.BillingAddress
    FROM Customer c
    JOIN Invoice i ON c.CustomerId = i.CustomerId
    WHERE c.Country = 'Canada'
    """, conn
)
```

### Task 3: Write a query to return the billing totals by country, in descending order of the totals.

```python
# call_on_students(1)
pd.read_sql(
    """
    SELECT c.Country, SUM(i.Total) AS Billing_Total
    FROM Customer c
    JOIN Invoice i ON c.CustomerId = i.CustomerId
    GROUP BY c.Country
    ORDER BY Billing_Total DESC;
    """, conn
)
```

## Part 2: Bayesian Statistics

$$ P(A|B) = \dfrac{P(B|A)P(A)}{P(B)}$$

### The Prompt: 
> In the second session of the 98th congress in 1984, a congressman voted 'No' on providing aid to El Salvador. 
> 
> Given that 60% of the congress were Democrats, 75% of whom voted 'No' for providing aid to El Salvador, and only 5% of Republicans voted 'No' to the proposal, what is the conditional probability that this individual is a Democrat?

### Task 5: Calculate the probability

```python
# call_on_students(1)
"""
P(Demdividual who voted 'No' on providing aid to El Salvador is a Democrat is approximately 95.7%
"""
```

**1. Which probability are we trying to find?**

   - The probability we are trying to find is the conditional probability that the individual who voted 'No' on providing aid to El Salvador is a Democrat.
   

**2. Based on that, what other pieces do we need?**

   - The probability that a randomly selected congressman is a Democrat (P(Democrat)).
   - The probability that a randomly selected congressman voted 'No' on providing aid to El Salvador given that they are a Democrat (P(No | Democrat)).
   - The probability that a randomly selected congressman voted 'No' on providing aid to El Salvador given that they are a Republican (P(No | Republican)).
    

**3. Result?**

   - P(Democrat) = 0.60 (60% of the congress were Democrats)
   - P(No | Democrat) = 0.75 (75% of Democrats voted 'No')
   - P(No | Republican) = 0.05 (5% of Republicans voted 'No')
*
   - P(Democrat | No) = (0.75 * 0.60) / (0.75 * 0.60 + 0.05 * 0.40)
   - P(Democrat | No) = 0.45 / (0.45 + 0.02)
   - P(Democrat | No) = 0.45 / 0.47
   - P(Democrat | No) ≈ 0.957 (approximately 95.7%)

### Task 8: Calculate a 95% confidence interval for the amount of sleep for one Freshman who sleeps 7 times with a mean of 5 hours and a standard deviation of 0.5 hours

```python
# call_on_students(1)
```

```python
# Code here
import scipy.stats as stats

mean = 5
standard_deviation = 0.5
n = 7
confidence_level = 0.95

standard_error = standard_deviation / (n ** 0.5)
margin_of_error = stats.t.ppf((1 + confidence_level) / 2, n - 1) * standard_error

lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error
print(lower_bound)
print(upper_bound)
```

### Task 9: Interpret the 95% Confidence Interval that you derived in the above question

```python
# call_on_students(1)
```

Interpret: 

We can be 95% confident that the true mean amount of sleep for one Freshman who sleeps 7 times falls within the interval of 4.537 to 5.318 hours

```python
# Run this cell without changes to set up for the next set of questions

# Grabbing the median car weight
weight_median = df['weight'].median()

# Separating based on light cars and heavy cars
heavy_cars = df.loc[df['weight'] > weight_median].sample(n=100, random_state=42)
light_cars = df.loc[df['weight'] <= weight_median].sample(n=100, random_state=42)

# Plotting histograms for acceleration for both light and heavy cars
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

ax1.set_title('Sample of Heavy Cars')
ax1.set_xlabel('Acceleration')
ax1.set_ylabel('Frequency')
ax1.hist(heavy_cars['acceleration'], bins=12)

ax2.set_title('Sample of Light Cars')
ax2.set_xlabel('Acceleration')
ax2.set_ylabel('Frequency')
ax2.hist(light_cars['acceleration'], bins=12)
plt.show()
```

### The Prompt: 

> Car manufacturers want to know if light cars accelerate faster than heavy cars.

### Task 11: Set up the null and alternative hypotheses for this test

```python
# call_on_students(1)
```

**Null hypothesis:**

- The mean acceleration of light cars is equal to the mean acceleration of heavy cars. (μ_light = μ_heavy)


**Alternative hypothesis:**

- The mean acceleration of light cars is greater than the mean acceleration of heavy cars. (μ_light > μ_heavy)


### Task 12: Describe your Type 1 and Type 2 errors in this context

```python
# call_on_students(1)
```

**Type I Error:**

- Concluding that light cars accelerate faster than heavy cars when, in reality, there is no significant difference in the mean acceleration between the two groups.


**Type II Error:**

- Failing to conclude that light cars accelerate faster than heavy cars when, in reality, there is a significant difference in the mean acceleration between the two groups.
